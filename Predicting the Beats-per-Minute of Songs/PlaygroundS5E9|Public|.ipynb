```python
def missing_data(data):
    total = data.isnull().sum()
    percent = (data.isnull().sum()/data.isnull().count()*100)
    tt = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
    types = []
    for col in data.columns:
        dtype = str(data[col].dtype)
        types.append(dtype)
    tt['Types'] = types
    return(np.transpose(tt))
```

```python
##Most frequent value
def most_frequent_values(data):
  total = data.count()
  tt = pd.DataFrame(total)
  tt.columns = ['Total']
  items = []
  vals = []
  for col in data.columns:
    try:
      item = data[col].value_counts().idxmax()
      val = data[col].value_counts().max()
      items.append(item)
      vals.append(val)
    except Exception as ex:
      print(ex)
      items.append(0)
      vals.append(0)
  tt['Most frequent items'] = items
  tt['Frequency'] = vals
  tt['Frequency %'] = tt['Frequency']/tt['Total']*100
  return(np.transpose(tt))
```

```python
most_frequent_values(train_df).T
```
